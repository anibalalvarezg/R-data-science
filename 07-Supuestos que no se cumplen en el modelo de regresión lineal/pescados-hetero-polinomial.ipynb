```r
data.fish<-read.csv('Fish.csv',sep=",")
View(data.fish)

data.fish2 <- within(data.fish, {Species <- factor(data.fish$Species, 
                                                   labels = c("Bream", "Roach","Whitefish","Parkki","Perch","Pike","Smelt"))})

plot(data.fish2$Height,data.fish2$Weight)
```

```r
##########################################################
#Estudiamos la relación altura vs. peso para cada especia#
##########################################################

##especia Bream
Bream.fish<-subset(data.fish2, Species=="Bream", select = c("Weight","Height"))
plot(Bream.fish$Height,Bream.fish$Weight)#relacion lineal clara, no hay heteroscedasticidad

Bream.poly<-lm(Bream.fish$Weight ~ poly(Bream.fish$Height, 2, raw=TRUE))
summary(Bream.poly)#Problema de precisión (colinealidad posible)

Bream.lin<-lm(Bream.fish$Weight~Bream.fish$Height)
summary(Bream.lin)

plot(Bream.fish$Height,Bream.fish$Weight)
abline(lm(Bream.fish$Weight~Bream.fish$Height), col = "red")#El ajuste lineal parece correcto
```

```r

##especia Roach
Roach.fish<-subset(data.fish2, Species=="Roach", select = c("Weight","Height"))
plot(Roach.fish$Height,Roach.fish$Weight)#relacion cuadratica? 

Roach.poly<-lm(Roach.fish$Weight ~ poly(Roach.fish$Height, 2, raw=TRUE))
summary(Roach.poly)

library(lmtest)
bptest(Roach.poly)#no hay heteroscedasticidad

plot(Roach.fish$Height,Roach.fish$Weight)
myPredict <- predict(Roach.poly) 
ix <- sort(Roach.fish$Height,index.return=T)$ix
lines(Roach.fish$Height[ix], myPredict[ix], col="red")
```

```r
##especia Whitefish
Whitefish.fish<-subset(data.fish2, Species=="Whitefish", select = c("Weight","Height"))
plot(Whitefish.fish$Height,Whitefish.fish$Weight)#relacion cuadratica clara

Whitefish.poly<-lm(Whitefish.fish$Weight ~ poly(Whitefish.fish$Height, 2, raw=TRUE))
summary(Whitefish.poly)
bptest(Whitefish.poly)#no hay heteroscedasticidad
#El estimador es insesgado, pero no el mejor. 
#De otra parte es verdad que el test de Student es erroneo, 
#pero podemos quedarnos seguros que el parametro del cuadrado es significativo

library(nlme)#intentemos una estimación más precisa
Whitefish.fish.gls1<-gls(Weight ~ poly(Height, 2, raw=TRUE),data=Whitefish.fish, weights = varPower()) #polinomio
summary(Whitefish.fish.gls1)
Whitefish.fish.gls2<-gls(Weight ~ poly(Height, 2, raw=TRUE),data=Whitefish.fish, weights = varExp()) #exponencial
summary(Whitefish.fish.gls2)
#El ajuste polinomial de la heteroscedasticidad parece mejor en vista de
#AIC, BIC, precision

plot(Whitefish.fish$Height,Whitefish.fish$Weight)
myPredict <- predict(Whitefish.fish.gls1) 
ix <- sort(Whitefish.fish$Height,index.return=T)$ix
lines(Whitefish.fish$Height[ix], myPredict[ix], col="red")

myPredict <- predict(Whitefish.poly) 
lines(Whitefish.fish$Height[ix], myPredict[ix], col="blue")
```

```r
##especia Parkki
Parkki.fish<-subset(data.fish2, Species=="Parkki", select = c("Weight","Height"))
plot(Parkki.fish$Height,Parkki.fish$Weight)#relacion cuadratica?

Parkki.poly<-lm(Parkki.fish$Weight ~ poly(Parkki.fish$Height, 2, raw=TRUE))
summary(Parkki.poly)
bptest(Parkki.poly)#hay heteroscedasticidad
```

```r
##########################################################
#Consideramos tests Student corregidos por metodo HC para#
#averiguar la significancia del termino cuadratico       #
##########################################################



library(sandwich)
#Se aplica los tests de Student robustos a la heteroscedasticidad
#con la function coeftest
coeftest(Parkki.poly, vcov. = vcovHC(Parkki.poly, type = "HC4"))
#No hay mas duda en que el termino cuadratico no es significativo

Parkki.lin<-lm(Parkki.fish$Weight~Parkki.fish$Height)
summary(Parkki.lin)
#Recordar que tenemos heteroscedasticidad. Por lo tanto:
#aunque el estimador es insesgado, pero no es el mejor. Ademas test de Student no valido

Parkki.fish.gls1<-gls(Weight ~ Height,data=Parkki.fish, weights = varPower())
summary(Parkki.fish.gls1)
Parkki.fish.gls2<-gls(Weight ~ Height,data=Parkki.fish, weights = varExp())
summary(Parkki.fish.gls2)
#Ajuste polinomial mejor de acuerdo al AIC, BIC y loglik
#Ajuste exponencial mejor de acuerdo a las desviaciones estandar

plot(Parkki.fish$Height,Parkki.fish$Weight)
lines(fitted(Parkki.fish.gls1) ~ Parkki.fish$Height, Parkki.fish, col = "red")
lines(fitted(Parkki.lin) ~ Parkki.fish$Height, Parkki.fish, col = "blue")
#################
```

```r
##especia Perch
Perch.fish<-subset(data.fish2, Species=="Perch", select = c("Weight","Height"))
plot(Perch.fish$Height,Perch.fish$Weight)#Eliminamos un outlier (valor cero sospechoso)
Perch.fish<-Perch.fish[-6,]
plot(Perch.fish$Height,Perch.fish$Weight)#relacion cuadratica?

Perch.poly<-lm(Perch.fish$Weight ~ poly(Perch.fish$Height, 2, raw=TRUE))
summary(Perch.poly)
bptest(Perch.poly)#no hay heteroscedasticidad

plot(Perch.fish$Height,Perch.fish$Weight)
myPredict <- predict(Perch.poly) 
ix <- sort(Perch.fish$Height,index.return=T)$ix
lines(Perch.fish$Height[ix], myPredict[ix], col="red")
```

```r
##especia Pike
Pike.fish<-subset(data.fish2, Species=="Pike", select = c("Weight","Height"))
plot(Pike.fish$Height,Pike.fish$Weight)

Pike.poly<-lm(Pike.fish$Weight ~ poly(Pike.fish$Height, 2, raw=TRUE))
summary(Pike.poly)#Estamos al limite por el termino cuadratico
bptest(Pike.poly)#no hay heteroscedasticidad, podemos confiar en lo anterior

Pike.lin<-lm(Pike.fish$Weight ~ Pike.fish$Height)
summary(Pike.lin)
#Segun el R^2 ajustado el ajuste polinomial es mejor

plot(Pike.fish$Height,Pike.fish$Weight)
myPredict <- predict(Pike.poly) 
ix <- sort(Pike.fish$Height,index.return=T)$ix
lines(Pike.fish$Height[ix], myPredict[ix], col="red")
```
